```python
import pandas as pd
from datetime import date
```

```python


class atp:


    def __init__(self, type_data):
        
        '''
        type_data: 'atp','doubles', 'futures','atp_players', 'rankings'
        
        '''

        BASE_URL = f'https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/'
        self.type_data = type_data
        todays_date = date.today() 
        current_year = todays_date.year 
        
        main_dict = {} 
        BASE_URL = f'https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/' 
        main_dict = {'atp': [],
                        'doubles':[] ,
                        'futures':[], 
                             'atp_players': f'{BASE_URL}atp_players.csv',
                            'rankings': [] 

                }
        yy_list = [str(i) for i in list(range(10,100, 10)) + ['00']] 
        for year in range(1968,current_year): 
            atp_temp = f'{BASE_URL}atp_matches_{year}.csv' ; main_dict['atp'].append(atp_temp)
            doubles_temp = f'{BASE_URL}atp_matches_doubles_{year}.csv' ; main_dict['doubles'].append(doubles_temp)
            futures_temp = f'{BASE_URL}atp_matches_futures_{year}.csv' ; main_dict['futures'].append(futures_temp)

        for yy in yy_list:
            try:
                rankings_temp =  f'{BASE_URL}atp_rankings_{yy}s.csv' ; main_dict['rankings'].append(rankings_temp)
            except:
                continue 

        self.URLS = main_dict
        #print (main_dict)
        li = []
        if isinstance(main_dict[type_data], str):
            list_files = [main_dict[type_data]]
        else:
            list_files = main_dict[type_data]
        for filename in list_files:
            print (f'Reading: {filename}')
            try:
                df = pd.read_csv(filename)
                li.append(df)
            except:
                continue

        self.data = pd.concat(li, axis=0, ignore_index=True)
    
        #self.data = {case:pd.read_csv(url) for case, url in self.URLS.items()}

    # create other useful functions to work with data
    def current_status(self):
        # function to show current status
        pass 
```
